```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, train_test_split
```

```python
# https://archive.ics.uci.edu/ml/datasets/seeds#
df = pd.read_csv('seeds_dataset.csv', sep='\s+', names=["Area", "Perimeter", "Compactness",
                                                        "Length", "Width", "Assymetry",
                                                        "Length_groove", "Class"])
df.describe()
```

```python
df.groupby('Class').mean()
```

```python
df.corr()
```

```python
features = df[df.columns[df.columns != 'Class']]
labels = df['Class']

# Классификатор по K соседним
classifier = KNeighborsClassifier(n_neighbors=4)
means = []

# KFold - дробитель данных для кросс-валидации
kf = KFold(n_splits=20)

for train,test in kf.split(features):
    X_train, X_test, y_train, y_test = features.loc[train], features.loc[test], labels[train], labels[test]
    classifier.fit(X_train, y_train)
    prediction = classifier.predict(features.loc[test])
    curmean = np.mean(prediction == labels[test])
    means.append(curmean)

print(np.mean(means))
```
